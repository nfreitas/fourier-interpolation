```python
%matplotlib notebook
import matplotlib.pyplot as plt

import numpy as np
import lfi

# test data
A = lfi.test_data()
plt.imshow(np.real(A))
```

```python
# regular fourier interpolation (it takes a while)
F, B = lfi.fi(A, factor=4)

plt.figure()
plt.imshow(np.real(B))
```

```python
# memory efficient fourier interpolation
# only computes the band from column 4000 to 6000

OUT = lfi.large_fi(A, factor=4, band=np.arange(4000,6000))

plt.figure()
plt.imshow(np.real(OUT))
```
